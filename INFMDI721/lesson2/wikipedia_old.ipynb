```python
def next_wiki_url(url, target="Philosophy"):
    got_it = False
    
    content = requests.get(url)
    soup = BeautifulSoup(content.text, 'html.parser')

    paragraphs = soup.select("p")

    title = soup.find('title')
    
    if target in title:
        got_it = True

    for p in paragraphs:
        links = p.select('a')
        for link in links:
            wiki_url = link.attrs['href']
            break;
    
    return (wiki_url, got_it)
```

```python
url = "https://en.wikipedia.org/wiki/Mathematics"

counter = 0
while True:
    (tmp, is_reached) = next_wiki_url(url)
    print("url = {}".format(tmp))
    if is_reached:
        break
    counter += 1
    
print("counter = {}", counter)
```
